```python
def solve(board, length) :
    end = len(board[0]) - 1
    ret = 0
    direct = [[0, end + 1, 1], [end, -1, -1]]
    for row in range(len(board)) :
        num = board[row][0]
        len_board = 0
        is_back = 0
        if board[row][0] > board[row][-1] :
            is_back = 1
            num = board[row][end]
        for col in range(direct[is_back][0], direct[is_back][1], direct[is_back][2]) :
            if num == board[row][col] :
                len_board += 1
            elif num + 1 == board[row][col] :
                if len_board >= length :
                    num = board[row][col]
                    len_board = 1
                else :
                    ret -= 1
                    #print(row)
                    break
            elif num - 1 == board[row][col]:
                is_break = False
                if col + (length-1) * direct[is_back][2] < 0 or col + (length-1) * direct[is_back][2] > end :
                    ret -= 1
                    break
                for temp_col in range(col, col + length * direct[is_back][2], direct[is_back][2]) :
                    if num - 1 != board[row][temp_col] :
                        ret -= 1
                        is_break = True
                        break
                if is_break == True :
                    break
                num -= 1
                len_board = 1 - length
            else :
                ret -=1
                #print(row)
                break
        ret += 1
    return ret

def solve2(board, length) :
    end = len(board[0]) - 1
    ret = 0
    direct = [[0, end + 1, 1], [end, -1, -1]]
    for col in range(len(board)) :
        num = board[0][col]
        len_board = 0
        is_back = 0
        if board[0][col] > board[-1][col] :
            is_back = 1
            num = board[end][col]
        for row in range(direct[is_back][0], direct[is_back][1], direct[is_back][2]) :
            if num == board[row][col] :
                len_board += 1
            elif num + 1 == board[row][col] :
                if len_board >= length :
                    num += 1
                    len_board = 1
                else :
                    ret -= 1
                    break
            elif num - 1 == board[row][col]:
                is_break = False
                if row + (length-1) * direct[is_back][2] < 0 or row + (length-1) * direct[is_back][2] > end :
                    ret -= 1
                    break
                for temp_row in range(row, row + length * direct[is_back][2], direct[is_back][2]) :
                    if num - 1 != board[temp_row][col] :
                        ret -= 1
                        is_break = True
                        break
                if is_break == True :
                    break
                num -= 1
                len_board = 1 - length
            else :
                ret -=1
                break
        ret += 1
    return ret
        
    
def Input() :
    n, length = list(map(int, input().split()))
    board = []
    for i in range(n):
        board.append(list(map(int, input().split())))
    return length, board
    
def main() :
    #board = [[3,3,3,3,3,3], [2,3,3,3,3,3], [2,2,2,3,2,3], [1,1,1,2,2,2], [1,1,1,3,3,1], [1,1,2,3,3,2]]
    #board = [[3, 2, 1, 1, 2, 3], [3, 2, 2, 1, 2, 3], [3, 2, 2, 2, 3, 3], [3, 3, 3, 3, 3, 3], [3, 3, 3, 3, 2, 2], [3, 3, 3, 3, 2, 2]]
    #board = [[3, 2, 1, 1, 2, 3],[3, 2, 2, 1, 2, 3],[3, 2, 2, 2, 3, 3],[3, 3, 3, 3, 3, 3],[3, 3, 3, 3, 2, 2],[3, 3, 3, 3, 2, 2]]
    #length = 3
    length, board = Input()
    res = solve(board, length)
    #print(res)
    res += solve2(board, length)
    print(res)
              
if __name__ == '__main__' :
    main()
    
'''
6 2
3 3 3 3 3 3
2 3 3 3 3 3
2 2 2 3 2 3
1 1 1 2 2 2
1 1 1 3 3 1
1 1 2 3 3 2
'''
```
