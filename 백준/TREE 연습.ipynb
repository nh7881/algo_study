# TREE 연습

```python
class Tree :
    def __init__(self, Root) :
        self.root = Root
    def add_child(self, par, node) :
        if par.val < node.val :
            if par.right != None :
                self.add_child(par.right, node)
            else :
                par.right = node
        else :
            if par.left != None :
                self.add_child(par.left, node)
            else :
                par.left = node
    def max_length(self, node) :
        if node.left != None :
            ret = self.max_length(node.left) + 1
        else :
            ret = 0
        if node.right != None :
            ret = max(ret, self.max_length(node.right) + 1)
        return ret
    
    def pre_order(self, node) :
        print(node.val, end = ' -> ')
        if node.left != None :
            self.pre_order(node.left)
        if node.right != None :
            self.pre_order(node.right)
            
    def in_order(self, node) :
        if node.left != None :
            self.in_order(node.left)
        print(node.val, end = ' -> ')
        if node.right != None :
            self.in_order(node.right)
            
    def post_order(self, node) :
        if node.left != None :
            self.post_order(node.left)
        if node.right != None :
            self.post_order(node.right)
        print(node.val, end = ' -> ')
    
    def search(self, node, val) :
        if node.val == val :
            return node
        if node.left != None :
            ret = self.search(node.left, val)
            if ret != -1 :
                return ret
        if node.right != None :
            ret = self.search(node.right, val)
            if ret != -1 :
                return ret
        return -1
    
    def num_sub_tree(self, node) :
        ret = 0
        if node.left != None :
            ret += 1 + self.num_sub_tree(node.left)
        if node.right != None :
            ret += 1 + self.num_sub_tree(node.right)
        return ret
    
    def num_big_sub_tree(self, node) :
        ret = 0
        if node.right != None :
            ret += 1 + self.num_sub_tree(node.right)
        return ret
    
    def num_small_sub_tree(self, node) :
        ret = 0
        if node.left != None :
            ret += 1 + self.num_sub_tree(node.left)
        return ret

class Node :
    def __init__(self, val) :
        self.val = val
        self.left = None
        self.right = None
        
def is_nerd(tree, human) :
    tree.add_child(tree.root, human)
    
def main() :
    node = Node(7)
    tree = Tree(node)
    vals = [5, 10, 6, 3, 1, 4, 12, 11]
    for val in vals :
        node = Node(val)
        tree.add_child(tree.root, node)
    print(tree.max_length(tree.root))
    print(tree.max_length(tree.root.left))
    print(tree.max_length(tree.root.right))
    print('preorder')
    tree.pre_order(tree.root)
    print()
    print('inorder')
    tree.in_order(tree.root)
    print()
    print('postorder')
    tree.post_order(tree.root)
    print()
    print(tree.search(tree.root, 12).val)
    print(tree.num_small_sub_tree(tree.search(tree.root, 7)))
    print(tree.num_big_sub_tree(tree.search(tree.root, 7)))
    
    
if __name__ == '__main__' :
    main()
```
