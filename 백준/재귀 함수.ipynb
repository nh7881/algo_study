### 재귀함수

```python
# 1. 팩토리얼
ret = 1
for n in range(2, int(input()) + 1) :
    ret *= n
print(ret)
```

```python
# 2. 피보나치 수 5
def fibonachi(n) :
    if n == 0 :
        return 0
    if n == 1 :
        return 1
    return fibonachi(n - 1) + fibonachi(n - 2)
print(fibonachi(int(input())))
```

```python
# 3. 별 찍기 - 10
def star(row, col, n) :
    global arr
    if n == 1 :
        arr[row][col] = '*'
        return
    sub = int(n / 3)
    for plus_row in range(0, n, sub):
        for plus_col in range(0, n, sub):
            if not(plus_row == sub and plus_col == sub) :
                star(row + plus_row, col + plus_col, sub)

def init_arr(n) :
    global arr
    for row in range(n) :
        arr.append([' '] * n)
        
def Print() :
    global arr
    for arr_row in arr :
        for val in arr_row :
            print(val, end = '')
        print()
n = int(input())
arr = []
init_arr(n)
star(0, 0, n)
Print()
```

```python
# 4. 하노이 탑 이동 순서
# hanoi(n) = hanoi(n-1) + 1 + hanoi(n-1)
def hanoi(n, start, mid, goal) :
    if n == 1 :
        global move
        moves.append([start, goal])
        return 1
    ret = 0
    ret += hanoi(n - 1, start, goal, mid)
    ret += hanoi(1, start, mid, goal)
    ret += hanoi(n - 1, mid, start, goal)
    return ret

n = int(input())
moves = []
res = hanoi(n, 1, 2, 3)
print(res)
for move in moves :
    print(move[0], move[1])
```
